```python
van = pd.read_csv('van_sauna_pool_steam.csv', parse_dates=True, index_col=0)
```

```python
van = van.sample(frac=1).reset_index(drop=True)
```

```python
def check_format(arr):
    pattern = r'\d+ft2'
    for count, entry in enumerate(arr):
        if re.match(pattern, entry):
            return([True, count])
    return([False, False])
```

```python
prices = []
bedrooms = []
sqfts = []
longitude = []
name = []
latitude = []
bathrooms_level_two = []
streetAddress = []
country = []
locality = []
postal = []
region = []
housing_type = []
links = []
sauna = []

count = 0
error = 0

for count, url in enumerate(van['URL']):
    try:
        user_agent = random.choice(user_agents)
        header = {"User-Agent": user_agent}

        proxy = random.choice(proxies)
        response = get(url, headers=header, proxies=proxy)

        # define the html text
        page_html = BeautifulSoup(response.text, 'html.parser')

        # find price
        price = page_html.find('span', class_='price').text
        print(price)

        # find sqft
        p_tag = page_html.find("p", class_='attrgroup')
        p_tag_array = p_tag.text.split()

        # checks if p_tag.text.split() has ft2 in it, returns [False, False] if it doesnt and returns [True, 'index pos of where sqft resides'] 
        contains_ft2 = check_format(p_tag_array)

        if contains_ft2[0] == True:
            sqfts.append(p_tag_array[contains_ft2[1]])
            print(p_tag_array[contains_ft2[1]])
        else:
            sqfts.append(np.nan)

        posting_data = page_html.find(id="ld_posting_data").text.strip()

        # convert to json dict
        posting_data = json.loads(posting_data)

        prices.append(price)
        bedrooms.append(posting_data['numberOfBedrooms'])
        longitude.append(posting_data['longitude'])
        name.append(posting_data['name'])
        latitude.append(posting_data['latitude'])
        bathrooms_level_two.append(posting_data['numberOfBathroomsTotal'])
        streetAddress.append(posting_data['address']['streetAddress'])
        country.append(posting_data['address']['addressCountry'])
        locality.append(posting_data['address']['addressLocality'])
        postal.append(posting_data['address']['postalCode'])
        region.append(posting_data['address']['addressRegion'])
        housing_type.append(posting_data['@type'])
        links.append(url)
        sauna.append(van.iloc[count, 1])

    except:
        error += 1
        print(f" error #{error}: Error processing page: {url}")
        prices.append(np.nan)
        bedrooms.append(np.nan)
        sqfts.append(np.nan)
        longitude.append(np.nan)
        name.append(np.nan)
        latitude.append(np.nan)
        bathrooms_level_two.append(np.nan)
        streetAddress.append(np.nan)
        country.append(np.nan)
        locality.append(np.nan)
        postal.append(np.nan)
        region.append(np.nan)
        housing_type.append(np.nan)
        links.append(url)
        sauna.append(van.iloc[count, 1])
        error += 1

    count += 1

    if count == 38:
        count = 0
        sleep(random.uniform(1.8, 2.7))
```

```python
result = [list(x) for x in zip(prices, bedrooms, sqfts, longitude, name, latitude, bathrooms_level_two, streetAddress, country,locality,postal,region,housing_type,links, van['sauna'], van['pool'], van['steam'])]
second_level_data_scrape = pd.DataFrame(result, columns = ['price', 'bedrooms', 'sqfts', 'longitude', 'name', 'latitude', 'bathrooms_level_two', 'streetAddress', 'country', 'locality', 'postal', 'region', 'housing_type', 'links', 'sauna', 'pool', 'steam'])
```
